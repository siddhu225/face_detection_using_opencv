```python
#converting to grey scale image
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Displaying grayscale image
plt.imshow(image_gray, cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('/home/siddhu/Downloads/opencv-master/data/haarcascades/haarcascade_frontalface_alt2.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(image, scaleFactor = 1.2, minNeighbors = 5);
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 4)
```
